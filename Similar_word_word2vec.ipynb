```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim.models import Word2Vec
```

```python
data="In the shade of the house, in the sunshine of the riverbank near the boats, in the shade of the Sal-wood forest, in the shade of the fig tree is where Siddhartha grew up, the handsome son of the Brahman, the young falcon, together with his friend Govinda, son of a Brahman. The sun tanned his light shoulders by the banks of the river when bathing, performing the sacred ablutions, the sacred offerings. In the mango grove, shade poured into his black eyes, when playing as a boy, when his mother sang, when the sacred offerings were made, when his father, the scholar, taught him, when the wise men talked. For a long time, Siddhartha had been partaking in the discussions of the wise men, practising debate with Govinda, practising with Govinda the art of reflection, the service of meditation. He already knew how to speak the Om silently, the word of words, to speak it silently into himself while inhaling, to speak it silently out of himself while exhaling, with all the concentration of his soul, the forehead surrounded by the glow of the clear-thinking spirit. He already knew to feel Atman in the depths of his being, indestructible, one with the universe."
```

```python
lm=WordNetLemmatizer()
```

```python
text=re.sub("[^A-Za-z0-9\s\d]","",data)
text=text.lower()
text
```

```python
sentences=nltk.sent_tokenize(text)
sentences
```

```python
sentences=[nltk.word_tokenize(sentence) for sentence in sentences]
sentences
```

```python
for i in range(len(sentences)):
      sentences[i]=[word for word in sentences[i] if word not in stopwords.words("english")]
```

```python
sentences
```

```python
model=Word2Vec(sentences,min_count=1)
```

```python
similar=model.wv.most_similar("glow")
similar
```
